```python
#coding:utf8
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os

from skimage.measure import compare_psnr
#import pybm3d
filename = "noise.jpg"
gtname = "noise_clean.jpg"
img = cv2.imread(filename)
imggt = cv2.imread(gtname)

def plotResiduals(residuals,pos,title):
    plt.subplot(pos)
    plt.title(title)
    plt.imshow(residuals)

plt.title("org pic")
plt.imshow(img)
plt.show()

##----高斯滤波----##
plt.figure(figsize=(16,10))
img_gaussian= cv2.GaussianBlur(img,(5,5),0)
plt.subplot(341)
plt.title("gauss blur")
plt.imshow(img_gaussian)
residuals = img - img_gaussian
plotResiduals(residuals,342,"gauss blur residuals")
psnr=compare_psnr(imggt,img_gaussian) #计算图像的峰值信噪比
print("gauss blur psnr:%f"%psnr)
```

```python
import cv2 
import matplotlib.pyplot as  plt 
import numpy as np 
from skimage.measure import compare_psnr

filename='noise.jpg'
gtname='noise_clean.jpg'
img=cv2.imread(filename)
imggt=cv2.imread(gtname)

def Residual(residual,size,title):
    plt.subplot(size)
    plt.title(title)
    plt.imshow(residual)
```

```python
#https://blog.csdn.net/on2way/article/details/46828567
#高斯滤波
plt.figure(figsize=(16,10))
img_gaussian=cv2.GaussianBlur(img,(5,5),0)
#plt.subplot(341)
#plt.title('img_gaussian')
#plt.imshow(img_gaussian)
Residual(img_gaussian,341,'img_gaussian')
residuals=img-img_gaussian
Residual(residuals,342,'residuals')
##----中值滤波----##
img_medianblur= cv2.medianBlur(img,5)
Residual(img_medianblur,343,'img_medianblur')
residuals = img - img_medianblur
Residual(residuals,344,'residuals')
##----双边滤波----##
img_bilateral= cv2.bilateralFilter(img,10,40,40)
Residual(img_bilateral,345,'img_bilateral')
residuals=img-img_bilateral
Residual(residuals,346,'residuals')
##----非局部均值滤波----##
img_nonlocal= cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)
Residual(img_nonlocal,347,'img_nonlocal')
residuals=img-img_nonlocal
Residual(residuals,348,'residuals')
```
