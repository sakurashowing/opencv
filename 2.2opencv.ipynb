```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img=np.array(Image.open('./æŸ¿å­.jpg').convert('L')) #æ¨¡å¼Lä¸ºç°è‰²å›¾åƒ
plt.figure("image")
arr=img.flatten()#æŠ˜å æˆä¸€ç»´çš„æ•°ç»„
n,bins,patches=plt.hist(arr,bins=256,alpha=0.75)
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
src=Image.open('./æŸ¿å­.jpg')
r,g,b=src.split()

plt.figure("lena")
ar=np.array(r).flatten()
plt.hist(ar, bins=256, normed=1,facecolor='r',edgecolor='r',hold=1)
ag=np.array(g).flatten()
plt.hist(ag, bins=256, normed=1, facecolor='g',edgecolor='g',hold=1)
ab=np.array(b).flatten()
plt.hist(ab, bins=256, normed=1, facecolor='b',edgecolor='b')
plt.show()
```

```python
#coding:utf8
import cv2
import matplotlib.pyplot as plt
import numpy as np

filename='shizi.jpg' #æŸ¿å­å½©è‰²å›¾

rgbimg=cv2.imread(filename)
rgbimg = cv2.resize(rgbimg,(600,400))
hsvimg = cv2.cvtColor(rgbimg,cv2.COLOR_BGR2HSV)
labimg = cv2.cvtColor(rgbimg,cv2.COLOR_BGR2Lab)

cv2.namedWindow("rgb",0)
cv2.namedWindow("lab",0)
cv2.namedWindow("hsv",0)

cv2.imshow("rgb",rgbimg)
cv2.imshow("hsv",hsvimg)
cv2.imshow("lab",labimg)

cv2.imwrite("æŸ¿å­rgb.jpg",rgbimg)
cv2.imwrite("æŸ¿å­hsv.jpg",hsvimg)
cv2.imwrite("æŸ¿å­lab.jpg",labimg)

cv2.waitKey(0)
```
